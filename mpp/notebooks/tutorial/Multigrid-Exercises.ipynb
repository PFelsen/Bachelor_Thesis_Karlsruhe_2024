Folgender Quellcode erstellt eine Tabelle f√ºr den Vergleich von Lagrange und DG-Diskretisierung. 

````python
mpp.reset_data()
models = ("LagrangeElliptic", "DGElliptic")
levels = (1,2,3,4)
for model, level in product(models, levels):
    data = {"Problem":"LaplaceSquare500", "level":level, "Model":model}
    mpp.run(4, config="TUT_elliptic", kwargs=data)
    mpp.parse_json()
````

````python
columns = [
    "Assemble Info.Assemble Name",
    "Mesh Info.Level", 
    "Mixed Info.DoFCount",           
    "Mixed Info.Goal Functional",
    "Mixed Info.Flux Loss",
    "Mixed Info.LinearSteps",
    "Computation Time"
]
pd.DataFrame(pd.json_normalize(mpp.data), columns=columns)
````
