### Exercise 4 (Subsurface flows)

In this exercise we want to look at subsurface flows modeled by
$$
\text{div}(\kappa(x,y)u(x,y)) = 0
$$
with the boundary conditions
$$
\begin{align*}
u(x,y) &= 0, \quad y = 0 \\
\nabla u(x,y) \cdot n &= -1, \quad y = 1 \\
\nabla u(x,y) \cdot n &= 0, \quad x \in \{0,1\}
\end{align*}
$$

````python
import sys
sys.path.append('../..')

import matplotlib.pyplot as plt
import python.mppy as mppy
from plotting import *
import pandas as pd

mpp = mppy.Mpp(kernels=4,
               mute=False,
               executable='M++',
               project_name='tutorial',
               cmake_args = ['-DBUILD_TUTORIAL=ON'])

mpp.clean_data()
mpp.build()
````

### Exercise 6 (Mixed Finite Elements)

The subsurface flow problem can also be modeled by
\begin{align*}
\text{div} q(x, y) &= 0, \quad (x,y) \in \Omega = (0, 1)^2 \\
    - \kappa \nabla u(x,y) &= q, \quad (x,y) \in \Omega =  (0, 1)^2
\end{align*}
with boundary conditions such as in Exercise 4. We want to use mixed finite elements to solve the resulting saddle point problem. After that we can calculate the so-called 'Dual-Primal Error'.

````python
mpp.reset_data()
mpp.run(4, config="TUT_mixed")
perm_solution_plot(mpp)
````

````python
mpp.reset_data()

columns = [
    "Mesh Info.Level", 
    "Mixed Info.Calculated Inflow", 
    "Mixed Info.Calculated Outflow",
    "Mixed Info.Flux Loss", 
    "Exact Solution Info.Flux Error", 
    "Computation Time"
]

kwargs = {"level": 2,"Problem": "LaplaceSquare500"}
mpp.run(4, config="TUT_mixed", kwargs=kwargs)

mpp.parse_json()
pd.DataFrame(pd.json_normalize(mpp.data), columns=columns)
````

````python
perm_solution_plot(mpp)
````

### Exercise 8 (Programming exercise - Rock problem)

In this exercise we consider $\Omega = (0,3)^2$ with a rock placed in $(1,2) \times (0,1) $ and an additional layer of clay in $(0,1)\times (0,1)$. The geometry and the boundary conditions can be found in 'tutorial/conf/geo/Rock.geo'. To specify the permeability function of the class Rock is defined in 'tutorial/src/problem/EllipticProblem.hpp' and in 'tutorial/src/assemble/EllipticAssemble.C' the function OutFlowLeftRight provides a way to calculate the outflow on the right side and the outflow on the left side of the rock seperatly.

````python
mpp.reset_data()

kwargs={"Model": "MixedElliptic", "level": 4, "Problem": "Rock",
        "Permeability": 0.1}

mpp.run(4, config="TUT_mixed", kwargs=kwargs)
perm_solution_plot(mpp)
````

````python
columns = [
    "Mesh Info.Level", 
    "Config Info.Permeability", 
    "Rock Problem Info.Left Outflow",
    "Rock Problem Info.Right Outflow", 
    "Dual Primal Error Info.Dual Primal Error", 
    "Computation Time"
]

kwargs={"Model": "MixedElliptic", "level": 4, "Problem": "Rock"}

for permeability in [0.5, 1.0, 2.5, 5.0]: 
    kwargs["Permeability"] = permeability
    mpp.run(4, config="TUT_mixed", kwargs=kwargs)
    mpp.parse_json()
    
perm_solution_plot(mpp)
pd.DataFrame(pd.json_normalize(mpp.data),columns=columns)
````

### Exercise 9 (Different discretization)

Through running, parsing and comparing different discretizations we can get a feeling for the bevahior of the different methods. Instead of ```config="laplace"``` we also can use the more general ```config="elliptic"```.

````python
# Todo: make this a meaningful experiment and remove warnings and errors

mpp.mute = False
mpp.reset_data()

columns = [
    "Config Info.Model", 
    "Mesh Info.Level", 
    "Config Info.degree", 
    "Mixed Info.Calculated Inflow", 
    "Mixed Info.Calculated Outflow",
    "Mixed Info.Goal Functional", 
    "Mixed Info.Flux Loss", 
    "Exact Solution Info.Flux Error", 
    "Computation Time"
]

# segmentation fault bei HybridElliptic
#models = ["LagrangeElliptic", "MixedElliptic", "HybridElliptic", "DGElliptic", "EGElliptic"] 
models = ["LagrangeElliptic", "MixedElliptic", "DGElliptic", "EGElliptic"]

for model in models:
    for level in range(2, 3):
        mpp.run(4, config="TUT_elliptic", kwargs={"Model": model, "level": level, "degree":1})
        mpp.parse_json()
    
pd.DataFrame(pd.json_normalize(mpp.data), columns=columns)
````
