# Experiment Configuration

```python
print(socket.gethostname())

if socket.gethostname().find("niklas") != -1:
    mpp.kernels = 4
    mpp.mute = False
    kwargs = {"epsilon":     0.0,
              "initLevels":  [4, 5, 6, 7],
              "initSamples": [20, 20, 20, 20]}
    epsilons = [0.05, 0.03, 0.01]
    
elif socket.gethostname().find("runner" ) != -1:
    mpp.kernels = 16
    mpp.mute = False
    kwargs = {"epsilon":     0.0,
              "initLevels":  [4, 5, 6, 7],
              "initSamples": [50, 50, 50, 50]}
    epsilons = [0.05, 0.03, 0.01, 0.005]
    
else:
    mpp.kernels = 64
    mpp.mute = False
    kwargs = {"epsilon":     0.0,
              "initLevels":  [3, 4, 5, 6, 7, 8],
              "initSamples": [500, 500, 500, 500, 500, 500]}
    epsilons = [0.05, 0.03, 0.01, 0.005, 0.003, 0.001, 0.0005]
```

## Convergence tests

```python
mpp.reset_data()

sigmas = [1.0, 1.5, 2.0]

for sigma in sigmas:
    kwargs["sigma"] = sigma
    mpp.run(config="mlmc_transport", kwargs=kwargs)
    mpp.parse_log()
    
convergence_plot(mpp, label="sigma")
cost_plot(mpp)
```

### Multilevel Monte Carlo Experiment

```python
mpp.reset_data()

kwargs={}

for epsilon in epsilons:
    kwargs["epsilon"] = epsilon
    mpp.run(config="mlmc_transport", kwargs=kwargs)
    mpp.parse_log()

convergence_plot(mpp, label="Epsilon")
value_error_plot(mpp)
mlmc_plot(mpp)
cost_plot(mpp)
```
